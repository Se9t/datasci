```python
mdf=df.drop(['Area Code','Area','Item Code','latitude','longitude','Element Code','Unit'],axis=1)
```

```python
pd.pivot_table(mdf,index=['Area Abbreviation'],columns='Element',aggfunc='sum',margins=True).tail()
# 数据总览
```

```python
mdf.groupby('Element').sum().sum(axis=1).plot.bar()
# 全世界1961到2013年饲料与粮食总产值 可见粮食大概是饲料的三倍
```

```python
mdf['all']=mdf.sum(axis=1)
```

```python
topitems=mdf.groupby(['Item'])['all'].sum()
topitems=topitems.to_frame().assign(pct=topitems/topitems.sum())
topitems.sort_values('pct',ascending=False)['pct'][:10].plot.pie(autopct='%0.2f%%')
# 全球产量前10的产品  谷物牛奶最多其次是蔬菜
```

```python
topn=mdf[['Area Abbreviation','all']].groupby('Area Abbreviation').sum().sort_values('all',ascending=False)
topn.assign(pct=topn/topn.sum())['pct'][:10].plot.pie(autopct='%0.2f%%')
# 全球产量前10的国家,中国美国印度占据全球40%多
```

```python
ts=mdf.groupby('Element').sum().T
ts.index=ts.index.str.lstrip('Y')
ts.drop('all',inplace=True)
```

```python
ts.head()
```

```python
ts.plot(xticks=[0,10,20,30,50])
# 从时间轴粮食产量在1991发生转折迅速增加并持续上涨
```

```python
area1=mdf.groupby('Area Abbreviation').sum().sort_values('all')[:5]
area2=mdf.groupby('Area Abbreviation').sum().sort_values('all')[-5:]
```

```python
area1=area1.T
area2=area2.T
area1.index=area1.index.str.lstrip('Y')
area2.index=area2.index.str.lstrip('Y')
area1.drop('all',inplace=True)
area2.drop('all',inplace=True)
```

```python
area1.plot(xticks=[0,10,20,30,50])
area2.plot(xticks=[0,10,20,30,50])
# 粮食产量最低5个国家和最高的5个国家可以明显看出91年后中国的产量明显增加
# 印度排第二而美国增速平缓甚至在2011前有所下降 其他国家增速都比较平缓可以从侧面看出中国的发展速度非常迅猛
```

```python
lst=mdf.columns[3:-1].tolist()
```

```python
top5=[]
for i in lst:
    x=mdf.groupby('Item')[i].sum().sort_values(ascending=False)[:5]
    top5.append(x)
```

```python
pd.DataFrame(top5).fillna(0).plot(xticks=[0,10,20,30,50],figsize=(9,6))
# 1961年到2013年全球前5的粮食产量变化
# 由此可见谷物和牛奶比较稳定，玉米，蔬菜，小麦等都有较大波动 也许人类的饮食习惯几十年间几乎没有什么变化
```
